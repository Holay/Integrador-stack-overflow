```python
import time
from sentence_transformers import SentenceTransformer
from openai import OpenAI

from transformers import AutoModel, AutoTokenizer
import torch
```

```python
# Inicializar los modelos
model_openai = "text-embedding-3-small"
model_sentence_transformer = SentenceTransformer('all-MiniLM-L6-v2')
client = OpenAI(api_key='')


# Definir las funciones de generación de embeddings
def get_openai_embedding(texts):
    response = client.embeddings.create(
        input=texts,
        model=model_openai
    ).data[0].embedding
    return response

def get_sentence_transformer_embedding(texts):
    return model_sentence_transformer.encode(texts)
```

# Validar que modelo entrega una similaridad Coseno menor, tomando 100 preguntas aleatorias.

Añadimos una columna con sus embeddings respectivos

```python
# Preparar los datos
data2 = data[['QuestionBody','AnswerOwnerUserId']][:2000]
```

```python
data2['OpenAi_embedding'] = data2.QuestionBody.apply(lambda x: get_openai_embedding(x))
```

```python
data2['Sentence_embedding'] = data2.QuestionBody.apply(lambda x: get_sentence_transformer_embedding(x))
```

```python
#data2['Gemini_embedding'] = data2.QuestionBody.apply(lambda x: generate_embeddings_gemini(x))

# Generar embeddings para las primeras 10 filas
data2.loc[:10, 'Gemini_embedding'] = data2.loc[:10, 'QuestionBody'].apply(lambda x: generate_embeddings_gemini(x))
```

```python
# Procesar datos en lotes de 100 registros
batch_size = 100

for start in range(10, 2000, batch_size):
    end = min(start + batch_size, 2000)
    for i in range(start, end):
        try:
            if len(data2.loc[i, 'QuestionBody']) <= 10000:  # Verificar el tamaño del texto
                data2.at[i, 'Gemini_embedding'] = generate_embeddings_gemini(data2.loc[i, 'QuestionBody'])
            else:
                print(f"Skipping index {i} due to text size limit.")
        except Exception as e:
            print(f"Error at index {i}: {e}")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Calcular la distancia de similaridad (distancia coseno)
def cosine_similarity(embedding1, embedding2):
    return 1 - cosine(embedding1, embedding2)
```

```python
# Function to find similar questions
def find_similar_questions(query_embedding, embeddings, top_k=5):
    similarities = cosine_similarity([query_embedding], embeddings)
    similar_indices = np.argsort(-similarities[0])[:1]
    return similar_indices
```

```python
new_question = "How do I center a div"

# Generar embeddings para realizar una prueba una pregunta
new_openai_embedding = get_openai_embedding([new_question])
new_sentence_transformer_embedding = get_sentence_transformer_embedding(new_question)
new_gemini_embedding = generate_embeddings_gemini(new_question)

# Calcular la similaridad con embeddings preexistentes
#similarities_sentence_transformer = [cosine_similarity(new_sentence_transformer_embedding, emb) for emb in sentence_transformer_embeddings]
similarities_sentence_transformer = [cosine_similarity(new_sentence_transformer_embedding, emb) for emb in np.array(data2['Sentence_embedding'].tolist())]
similarities_openai = [cosine_similarity(new_openai_embedding, emb) for emb in np.array(data2['OpenAi_embedding'].tolist())]
similarities_gemini = [cosine_similarity(new_gemini_embedding, np.array(data2['Gemini_embedding'].tolist()))]

# Obtener la mejor distancia de similaridad
best_similarity_sentence_transformer = max(similarities_sentence_transformer)
best_similarity_openai = max(similarities_openai)
best_similarity_gemini = max(similarities_gemini)

print("Mejor distancia de similaridad con Sentence Transformer:", best_similarity_sentence_transformer)
print("Mejor distancia de similaridad con OpenAI:", best_similarity_openai)
print("Mejor distancia de similaridad con Gemini:", best_similarity_gemini)

```

```python
import numpy as np
import random

# Extraer embeddings para las 100 preguntas aleatorias seleccionadas
selected_questions = data2.sample(n=100, random_state=42).reset_index(drop=True)

# Calcular cosine similarity entre embeddings
cosine_sim_openai = []
cosine_sim_sentence = []
cal_gemini_embedding = []

for i, row in selected_questions.iterrows():
    cal_openai_embedding = get_openai_embedding(row["QuestionBody"])
    cal_sentence_transformer_embedding = get_sentence_transformer_embedding(row["QuestionBody"])
    cal_gemini_embedding = generate_embeddings_gemini(row["QuestionBody"])

    cosine_sim_openai.append( [cosine_similarity(cal_openai_embedding, emb) for emb in np.array(data2['OpenAi_embedding'].tolist())])
    cosine_sim_sentence.append( [cosine_similarity(cal_sentence_transformer_embedding, emb) for emb in np.array(data2['Sentence_embedding'].tolist())])
    cal_gemini_embedding.append( [cosine_similarity(cal_gemini_embedding, emb) for emb in np.array(data2['Gemini_embedding'].tolist())])

# Analizar resultados
avg_cosine_sim_openai = np.mean(cosine_sim_openai)
avg_cosine_sim_sentence = np.mean(cosine_sim_sentence)
avg_cosine_sim_gemini = np.mean(cal_gemini_embedding)


print(f"Promedio distancia coseno para OpenAI (text-embedding-3-small): {avg_cosine_sim_openai:.2f}")
print(f"Promedio distancia coseno para  SentenceTransformer (all-MiniLM-L6-v2): {avg_cosine_sim_sentence:.2f}")
print(f"Promedio distancia coseno para  Gemini (text-embedding-004): {avg_cosine_sim_gemini:.2f}")

```

tomamos 100 preguntas aleatorias y calculamos la media con cada uno de los modelos para ver cual entrega el valor de los emmbedings mas bajo.
El resultado nos entrega que SentenceTransformer (all-MiniLM-L6-v2) tiene la similaridad coseno mas baja con 0.02. Lo que indica que tiene los resultados mas fiables
